```python
import numpy as np
from scipy import stats

data_a = np.random.normal(2.0,2.0,100)
# 평균이 2고 표준편차가 2인 정규분포에서 100개의 원소로 배열을 만듬
```

```python
data_a[:5]
```

```python
estimator = stats.gaussian_kde(data_a, bw_method='silverman')
x = np.arange(-10,10)
k = estimator(x)
plt.plot(x,k,label='gaussian_kde')
plt.show()
```
