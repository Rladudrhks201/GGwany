```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
def linear_function(x):
    a = 0.5
    b = 2

    return a*x+b
```

```python
print(linear_function(5))
```

```python
x = np.arange(-5,5,0.1)
y = linear_function(x)
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('LinearFunction')
plt.show()
```

```python
# 이차 선형회귀 곡선
def quadratic_function(x):
    a = 1
    b = -1
    c = -2

    return a*x**2 + b*x + c
```

```python
print(quadratic_function(2))
```

```python
# 삼차 선형회귀 곡선
def cubic_function(x):
    a = 4
    b = 0
    c = -1
    d = -8

    return a*x**3 + b*x**2 + c*x + d
```

```python
y = quadratic_function(x)
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('quadratic_function')
plt.show()
```

```python
y = cubic_function(x)
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('cubic_function')
plt.show()
```

```python
def my_func(x):
    a =1 
    b = -3
    c = 10

    return a*x**2 + b*x + c # 이차 함수의 수식
```

```python
x = np.arange(-10,10,0.1)
y = my_func(x)

plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('My_function')
plt.scatter(1.5,my_func(1.5))
plt.show()
```

```python
# 지수 함수 y = a^x
def exponential_function(x):
    a = 4
    return a**x
```

```python
print(exponential_function(4))
print(exponential_function(0))
```

```python
x = np.arange(-3,2,0.1)
y = exponential_function(x)
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('exponential_function')
plt.show()  
```
