```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(iris_dataset.data,iris_dataset.target,random_state=0,test_size=0.3)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
iris_dataframe = pd.DataFrame(X_train,columns=iris_dataset.feature_names)
```

```python
y_train
```

```python
pd.plotting.scatter_matrix(iris_dataframe,figsize=(15,15),marker='o',c=y_train,cmap='viridis')
plt.show()
```

```python
# 첫 번째 머신 러닝 모델: k-최근접 알고리즘
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
# 모델의 평가
knn.fit(X_train,y_train)
print(knn.score(X_train,y_train))
print(knn.score(X_test,y_test))
```

```python
import numpy as np

X_new = np.array([[5,2.9,1,0.2]]) #예측용 샘플 데이터
prediction = knn.predict(X_new)
print(prediction)
print(iris_dataset.target_names[0])
```

```python
# 모델의 예측값
y_pred = knn.predict(X_test)
print(y_pred)
print(np.mean(y_pred==y_test))
```
