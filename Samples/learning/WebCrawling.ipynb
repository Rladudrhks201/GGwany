```python
import requests
import bs4
```

```python
url = 'https://search.shopping.naver.com/search/category/100005307'
result = requests.get(url).text
result
```

```python
bsObj = bs4.BeautifulSoup(result,'html.parser')
```

```python
bsObj.find('a',{'class':'basicList_link__JLQJf'}).text
```

```python
bsObj.find_all('a',{'class':'basicList_link__JLQJf'})[1].text
```

```python
items = bsObj.find_all('a',{'class':'basicList_link__JLQJf'})
for item in items:
    print(item.text)
```

```python
for i in range(4):
    print(items[i].text)
```
